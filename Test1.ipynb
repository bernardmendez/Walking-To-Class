```python
points = pd.read_csv('track_points.csv', header =0)
```

```python
going = points.loc[:, ['X']]
```

```python
going.head()
```

```python
for gog in going['X']:
    degrees = int(gog)
    lob = (gog - degrees) * 60
    minutes = int(lob)
    seconds = round(((lob - minutes) * 60), 2)
    print(degrees, minutes, seconds)
```
